# Data Visualisation

```python
# Checking distribution of delays
delay_counts = df['Delay'].value_counts()
print("Delay distribution:\n", delay_counts)
```

```python
# Count of delays (0 = On Time, 1 = Delayed)
delay_counts = df['Delay'].value_counts().sort_index()

plt.figure(figsize=(6, 4))
ax = sns.barplot(x=delay_counts.index, y=delay_counts.values, palette='Set2')

# Rename X-axis labels
ax.set_xticklabels(['On Time', 'Delayed'])

# Y-axis label and title
plt.ylabel("Number of Flights")
plt.title("Flight Delay Distribution")

# Add count labels on top of bars
for i, count in enumerate(delay_counts.values):
    plt.text(i, count + 50, str(count), ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

### According to the data provided, approximately 70% of Southwest Airlines flights are delayed.

```python
#Airline-wise Delay Rate
airline_delay_rate = df.groupby("Airline")["Delay"].mean().sort_values(ascending=False)
print(airline_delay_rate)

airline_delay_rate.plot(kind='bar', figsize=(10,4), color='blue', title='Airline Delay Rate')
plt.ylabel('Delay Rate')
plt.grid(axis='y')
plt.show()
```

### Flights were delayed on various weekdays. Which day of the week is the safest for travel?

```python
# Delay by Day of Week
day_delay = df.groupby("DayOfWeek")["Delay"].mean()
day_delay.plot(kind='line', marker='o', title='Delay Rate by Day of Week')
plt.xticks(range(1, 8), ['Mon','Tue','Wed','Thu','Fri','Sat','Sun'])
plt.ylabel('Delay Rate')
plt.grid(True)
plt.show()
```

```python
# Delay by Time of Day
# Create bins: morning, afternoon, evening, night
bins = [0, 600, 1200, 1800, 2400]
labels = ['Early Morning', 'Morning', 'Afternoon', 'Evening']
df['TimeBin'] = pd.cut(df['Time'], bins=bins, labels=labels, right=False)

time_delay = df.groupby("TimeBin")["Delay"].mean()
time_delay.plot(kind='bar', title='Delay Rate by Time of Day', color='purple')
plt.ylabel('Delay Rate')
plt.grid(axis='y')
plt.show()
```

### Which airlines should be recommended for short-, medium-, and long-distance travel?

```python
# Create length categories
df['LengthCategory'] = pd.qcut(df['Length'], q=3, labels=['Short', 'Medium', 'Long'])

# Group by LengthCategory and Airline to get delay rate
airline_performance = df.groupby(['LengthCategory', 'Airline'])['Delay'].mean().reset_index()

# Sort within each length category
top_airlines_by_length = airline_performance.sort_values(['LengthCategory', 'Delay'])

# Display top 3 airlines per category
for category in ['Short', 'Medium', 'Long']:
    print(f"\n Recommended Airlines for {category}-distance flights:")
    print(top_airlines_by_length[top_airlines_by_length['LengthCategory'] == category].head(2))
```

 ### Do you notice any patterns in the departure times of long-duration flights?

```python
# Create flight duration categories
df['LengthCategory'] = pd.qcut(df['Length'], q=3, labels=['Short', 'Medium', 'Long'])

# Focus only on long-distance flights
long_flights = df[df['LengthCategory'] == 'Long']

# Create time-of-day bins
bins = [0, 600, 1200, 1800, 2400]
labels = ['Early Morning', 'Morning', 'Afternoon', 'Evening']
long_flights['TimeOfDay'] = pd.cut(long_flights['Time'], bins=bins, labels=labels, right=False)

# Count how many long-distance flights depart in each time slot
time_dist = long_flights['TimeOfDay'].value_counts().sort_index()

# Plot it
time_dist.plot(kind='bar', color='salmon', title='Departure Time Distribution of Long-Distance Flights')
plt.ylabel('Number of Flights')
plt.grid(axis='y')

# Add flight counts on top of the bars
for i, v in enumerate(time_dist):
    plt.text(i, v + 5, str(v), ha='center', fontweight='bold')

plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

###  How many flights were delayed at large hubs compared to medium hubs?

```python
# Filter only large and medium airports
filtered = df[df['type_source_airport'].isin(['large_airport', 'medium_airport'])]

# Group and count delayed flights per hub type
delay_by_hub = filtered[filtered['Delay'] == 1]['type_source_airport'].value_counts()

# Plotting
colors = ['tomato', 'gold']
delay_by_hub.plot(kind='bar', color=colors, title='Delayed Flights by Hub Type (Source Airport)')
plt.xlabel('Airport Type')
plt.ylabel('Number of Delayed Flights')
plt.xticks(rotation=0)
plt.grid(axis='y')

# Add actual counts on top of bars
for i, v in enumerate(delay_by_hub):
    plt.text(i, v + 5, str(v), ha='center', fontweight='bold')

plt.show()
```

```python
# Group and count delayed flights per hub type
delay_by_hub = filtered[filtered['Delay'] == 1]['type_dest_airport'].value_counts()

# Plotting
colors = ['tomato', 'gold']
delay_by_hub.plot(kind='bar', color=colors, title='Delayed Flights by Hub Type (Destination Airport)')
plt.xlabel('Airport Type')
plt.ylabel('Number of Delayed Flights')

# Add actual counts on top of bars
for i, v in enumerate(delay_by_hub):
    plt.text(i, v + 5, str(v), ha='center', fontweight='bold')

plt.show()
```

## Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
# Define feature set (X) and target variable (y)
X = df.drop('Delay', axis=1)
y = df['Delay']

# Perform the train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2,      # 20% for testing
    random_state=42,    # Reproducibility
    stratify=y          # Maintain class balance
)

# Print shapes to confirm
print(f"Training set size: {X_train.shape}")
print(f"Testing set size: {X_test.shape}")
```

```python
# Define features (X) and target (y)
X = df.drop('Delay', axis=1)
y = df['Delay']

# Select only numeric columns for standardization
numeric_features = X.select_dtypes(include='number').columns
X_numeric = X[numeric_features]

# Split into train and test sets
X_train_num, X_test_num, y_train, y_test = train_test_split(
    X_numeric, y, test_size=0.2, stratify=y, random_state=42
)

# Standardize the numeric features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_num)
X_test_scaled = scaler.transform(X_test_num)

# Convert scaled arrays back to DataFrames
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=numeric_features, index=X_train_num.index)
X_test_scaled_df = pd.DataFrame(X_test_scaled, columns=numeric_features, index=X_test_num.index)
```

```python
#Drop rows with missing target
df = df.dropna(subset=['Delay'])

#Features and target
X = df.drop('Delay', axis=1)
y = df['Delay']

#Select numeric features only
numeric_features = X.select_dtypes(include='number').columns
X_numeric = X[numeric_features]

#Handle missing numeric values by imputing mean
imputer = SimpleImputer(strategy='mean')
X_numeric_imputed = imputer.fit_transform(X_numeric)

#Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_numeric_imputed, y, test_size=0.2, stratify=y, random_state=42
)

#Standardize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

#Train Logistic Regression (SGD)
logreg_model = SGDClassifier(loss='log_loss', max_iter=1000, random_state=42)
logreg_model.fit(X_train_scaled, y_train)
y_pred = logreg_model.predict(X_test_scaled)

#Evaluate the model
print("Classification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")

print(confusion_matrix(y_test, y_pred))
```

```python
# Drop rows where target is missing
df = df.dropna(subset=['Delay'])

# Features and target
X = df.drop('Delay', axis=1)
y = df['Delay']

# Use only numeric features
numeric_features = X.select_dtypes(include='number').columns
X_numeric = X[numeric_features]

# Impute missing values with mean
imputer = SimpleImputer(strategy='mean')
X_numeric_imputed = imputer.fit_transform(X_numeric)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_numeric_imputed, y, test_size=0.2, stratify=y, random_state=42
)

# Standardize numeric features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train Decision Tree with controlled depth and minimum samples
tree_model = DecisionTreeClassifier(
    max_depth=5,             # Limit depth to avoid overfitting
    min_samples_split=10,    # Minimum samples required to split
    min_samples_leaf=5,      # Minimum samples in a leaf
    random_state=42
)
tree_model.fit(X_train_scaled, y_train)
y_pred = tree_model.predict(X_test_scaled)

# Evaluation
print("Decision Tree Classification Report:")
print(classification_report(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))


```

```python
# Train logistic regression using SGD
logreg_model = SGDClassifier(loss='log_loss', max_iter=1000, random_state=42)
logreg_model.fit(X_train_scaled, y_train)

# Logistic Regression predictions
y_train_pred_logreg = logreg_model.predict(X_train_scaled)
y_test_pred_logreg = logreg_model.predict(X_test_scaled)

# Decision Tree predictions
y_train_pred = tree_model.predict(X_train_scaled)
y_test_pred = tree_model.predict(X_test_scaled)

# Accuracy and reports
print("Logistic Regression - Train Accuracy:", accuracy_score(y_train, y_train_pred_logreg))
print("Logistic Regression - Test Accuracy:", accuracy_score(y_test, y_test_pred_logreg))
print("Classification Report (Logistic Regression):")
print(classification_report(y_test, y_test_pred_logreg))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
# Train Gradient Boosting Classifier
gb_model = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
gb_model.fit(X_train_scaled, y_train)
```

```python
# Predictions
y_train_pred_gb = gb_model.predict(X_train_scaled)
y_test_pred_gb = gb_model.predict(X_test_scaled)
# Accuracy
print("Gradient Boosting - Train Accuracy:", accuracy_score(y_train, y_train_pred_gb))
print("Gradient Boosting - Test Accuracy:", accuracy_score(y_test, y_test_pred_gb))
print("Classification Report (Gradient Boosting):")
print(classification_report(y_test, y_test_pred_gb))
```

```python
# Train Random Forest
rf_model = RandomForestClassifier(n_estimators=100, max_depth=8, random_state=42)
rf_model.fit(X_train_scaled, y_train)
y_train_pred_rf = rf_model.predict(X_train_scaled)
y_test_pred_rf = rf_model.predict(X_test_scaled)

# Accuracy & Reports
print("Random Forest - Train Accuracy:", accuracy_score(y_train, y_train_pred_rf))
print("Random Forest - Test Accuracy:", accuracy_score(y_test, y_test_pred_rf))
print("Classification Report (Random Forest):")
print(classification_report(y_test, y_test_pred_rf))
```

```python
# Train XGBoost
xgb_model = XGBClassifier(n_estimators=100, max_depth=6, learning_rate=0.1, 
                          use_label_encoder=False, eval_metric='logloss', random_state=42)
xgb_model.fit(X_train_scaled, y_train)
y_train_pred_xgb = xgb_model.predict(X_train_scaled)
y_test_pred_xgb = xgb_model.predict(X_test_scaled)

#Accuracy and Reports
print("\nXGBoost - Train Accuracy:", accuracy_score(y_train, y_train_pred_xgb))
print(" XGBoost - Test Accuracy:", accuracy_score(y_test, y_test_pred_xgb))
print("Classification Report (XGBoost):")
print(classification_report(y_test, y_test_pred_xgb))
```
